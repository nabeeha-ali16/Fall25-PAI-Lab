```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Upload the Input Image

```python
pic = cv2.imread("picture.PNG") 
```

```python
if pic is None:
    print("Error: image1 not loaded. Check the path!")
```

### Visualize Input Image

```python
pic1_rgb = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 6))
plt.imshow(pic1_rgb)
plt.title('Input Image 1')
plt.axis('off')
plt.show()
```

### Resize Image

```python
resized = cv2.resize(pic, (300, 300))

plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

### Crop Image

```python
cropped = pic[50:300, 50:300]   # adjust values if needed

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

### Convert to Grayscale

```python
gray = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

### Draw Shapes

```python
# Draw Line
line = pic.copy()
cv2.line(line, (50, 50), (300, 50), (0, 255, 0), 3)

plt.imshow(cv2.cvtColor(line, cv2.COLOR_BGR2RGB))
plt.title("Line Drawn")
plt.axis("off")
plt.show()
```

```python
# Draw Rectangle
rect = pic.copy()
cv2.rectangle(rect, (50, 50), (300, 300), (255, 0, 0), 3)

plt.imshow(cv2.cvtColor(rect, cv2.COLOR_BGR2RGB))
plt.title("Rectangle Drawn")
plt.axis("off")
plt.show()
```

```python
# Draw Circle

circle = pic.copy()
cv2.circle(circle, (200, 200), 100, (0, 0, 255), 3)

plt.imshow(cv2.cvtColor(circle, cv2.COLOR_BGR2RGB))
plt.title("Circle Drawn")
plt.axis("off")
plt.show()
```

### Add Text

```python
text = pic.copy()
cv2.putText(text, "Kites Cafe", (50, 50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (255, 255, 0), 2)

plt.imshow(cv2.cvtColor(text, cv2.COLOR_BGR2RGB))
plt.title("Text Added")
plt.axis("off")
plt.show()
```

### Thresholding

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap="gray")
plt.title("Threshold Image")
plt.axis("off")
plt.show()
```

### Edge Detection

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```

### Save Image

```python
cv2.imwrite("output.jpg", edges)
```
