## Water Jug Problem

```python
jug1_cap = 4
jug2_cap = 3
goal = 2

visited = set()

def dfs(state, path):
    x, y = state
    
    if x == goal or y == goal:
        path.append(state)
        print("Solution Found:")
        for i in range(len(path)-1):
            print(f"{path[i]} - {path[i+1]}")
        return True

    visited.add(state)
    path.append(state)

    next_states = []

    next_states.append((jug1_cap, y))
    
    next_states.append((x, jug2_cap))
    
    next_states.append((0, y))
    
    next_states.append((x, 0))
    
    transfer = min(x, jug2_cap - y)
    next_states.append((x - transfer, y + transfer))
    
    transfer = min(y, jug1_cap - x)
    next_states.append((x + transfer, y - transfer))


    for state_next in next_states:
        if state_next not in visited:
            if dfs(state_next, path.copy()):
                return True

    return False

initial_state = (0, 0)
dfs(initial_state, [])
```
