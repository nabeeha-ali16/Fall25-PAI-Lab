## House Price Prediction

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score, mean_squared_error
```

### Loading Dataset

```python
def load_data(train_file, test_file):
    train = pd.read_csv(train_file)
    test = pd.read_csv(test_file)

    print("Train Shape:", train.shape)
    print("Test Shape:", test.shape)

    return train, test
```

### Data Overview

```python
def explore_data(data):

    print("\nFirst 5 Rows")
    print(data.head())

    print("\nDataset Info")
    print(data.info())

    print("\nMissing Values")
    print(data.isnull().sum())
```

### Feature Selection

```python
def select_features(data):

    features = [
        "LotArea",
        "YearBuilt",
        "1stFlrSF",
        "2ndFlrSF",
        "FullBath",
        "BedroomAbvGr"
    ]

    X = data[features]
    y = data["SalePrice"]

    return X, y, features
```

### Handle Missing Values

```python
def handle_missing_values(X):

    X = X.fillna(X.mean())

    return X
```

### Split Dataset

```python
def split_data(X, y):

    X_train, X_valid, y_train, y_valid = train_test_split(
        X,
        y,
        train_size=0.8,
        test_size=0.2,
        random_state=42
    )

    return X_train, X_valid, y_train, y_valid
```

### Train Model

```python
def train_model(X_train, y_train):

    model = RandomForestRegressor(
        n_estimators=200,
        random_state=42
    )

    model.fit(X_train, y_train)

    return model
```

### Evaluate Model 

```python

def evaluate_model(model, X_valid, y_valid):

    predictions = model.predict(X_valid)

    r2 = r2_score(y_valid, predictions)
    mse = mean_squared_error(y_valid, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_valid, predictions)

    print("\nModel Evaluation")
    print("----------------------")
    print("R2 Score:", r2)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("MAE:", mae)
```

### Train on Full Data

```python
def train_full_model(model, X, y):

    model.fit(X, y)

    return model
```

### Predict Test Data

```python
def predict_test(model, test_data, features):

    X_test = test_data[features]

    X_test = X_test.fillna(X_test.mean())

    predictions = model.predict(X_test)

    return predictions
```

### Create Submission File

```python
def create_submission(test_data, predictions):

    submission = pd.DataFrame({
        "Id": test_data["Id"],
        "SalePrice": predictions
    })

    submission.to_csv("submission.csv", index=False)

    print("\nSubmission file created successfully!")
```

### Main Program

```python
train_data, test_data = load_data("train.csv", "test.csv")

explore_data(train_data)

X, y, features = select_features(train_data)

X = handle_missing_values(X)

X_train, X_valid, y_train, y_valid = split_data(X, y)

model = train_model(X_train, y_train)

evaluate_model(model, X_valid, y_valid)

model = train_full_model(model, X, y)

test_predictions = predict_test(model, test_data, features)

create_submission(test_data, test_predictions)
```
